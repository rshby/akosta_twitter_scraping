```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(context="talk")
from wordcloud import WordCloud
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
```

# Import Dataset

```python
df = pd.read_feather("dataset/data_sentiment.feather")
df
```

```python
df_negatif = df[df["sentiment_label"] == "negatif"].reset_index(drop=True)
df_positif = df[df["sentiment_label"] == "positif"].reset_index(drop=True)
```

# Gabungkan Text Menjadi Satu

```python
sentences_positif = ""
sentences_negatif = ""
for i in range(len(df_positif["kalimat_clean"])):
    kalimat = df_positif["kalimat_clean"][i]
    sentences_positif += kalimat + " "
    
for i in range(len(df_negatif["kalimat_clean"])):
    kalimat = df_negatif["kalimat_clean"][i]
    sentences_negatif += kalimat + " "
```

```python
sentences_negatif
```

```python
wc = WordCloud(
    width=1200,
    height=1200,
    background_color="white",
    colormap="Reds",
    stopwords=StopWordRemoverFactory().get_stop_words(),
    min_font_size=10
).generate(sentences_negatif)

plt.figure(figsize=(10, 10), constrained_layout=True, facecolor=None)
plt.imshow(wc)
plt.axis("off")
plt.tight_layout(pad=0)

plt.show()
```

```python
wc = WordCloud(
    width=1200,
    height=1200,
    background_color="white",
    colormap="Blues",
    stopwords=StopWordRemoverFactory().get_stop_words(),
    min_font_size=10
).generate(sentences_positif)

plt.figure(figsize=(10, 10), constrained_layout=True, facecolor=None)
plt.imshow(wc)
plt.axis("off")
plt.tight_layout(pad=0)

plt.show()
```
