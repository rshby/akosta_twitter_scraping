```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(context="talk")
import nltk
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
```

```python
df = pd.read_feather("dataMamiKostIndonesia.feather")
```

```python
df["text"]
```

```python
stopword_ind = StopWordRemoverFactory().get_stop_words()
```

```python
df['tokenized_sents'] = df.apply(lambda row: nltk.word_tokenize(row['text']), axis=1)
```

```python
df.loc[:, ["text", "tokenized_sents"]]
```

```python
def remove_stopword(tokenized):
    try:
        txt_clean = [word for word in tokenized if word not in stopword_ind and word.isalpha()]
        return txt_clean
    except Exception as e:
        print(f"kesalahan: {e}")
```

```python
df["clean_text"] = df["tokenized_sents"].apply(lambda x: remove_stopword(x))
```

## Words Cloud

```python
from wordcloud import WordCloud
```

```python
sentences = ""
for i in range(len(df["clean_text"])):
    kalimat = " ".join(df["clean_text"][i])
    sentences += kalimat + " "
```

```python
sentences = sentences.replace("rt", "")
sentences = sentences.replace("sbmptnfess", "")
sentences = sentences.replace("tegalfess", "")
sentences = sentences.replace("https", "")
sentences = sentences.replace("sbyfess", "")
sentences = sentences.replace("keniaos", "")
sentences = sentences.replace("wkwk", "")
sentences = sentences.replace("collegemenfess", "")
sentences = sentences.replace("que", "")
sentences = sentences.replace("melankos", "")
```

```python
sentences
```

```python
wc = WordCloud(
    width=1200,
    height=1200,
    background_color="white",
    stopwords=stopword_ind,
    min_font_size=10
).generate(sentences)

plt.figure(figsize=(10, 10), constrained_layout=True, facecolor=None)
plt.imshow(wc)
plt.axis("off")
plt.tight_layout(pad=0)

plt.show()
```
