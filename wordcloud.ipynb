```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(context="talk")
from wordcloud import WordCloud
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
```

# Import Dataset

```python
df = pd.read_feather("df_clean_indonesia.feather")
```

# Gabungkan clean_text menjadi satu kalimat

```python
sentences = ""
for i in range(len(df["clean_text"])):
    kalimat = " ".join(df["clean_text"][i])
    sentences += kalimat + " "
```

```python
sentences = sentences.replace("rt", "")
sentences = sentences.replace("sbmptnfess", "")
sentences = sentences.replace("tegalfess", "")
sentences = sentences.replace("https", "")
sentences = sentences.replace("sbyfess", "")
sentences = sentences.replace("keniaos", "")
sentences = sentences.replace("wkwk", "")
sentences = sentences.replace("collegemenfess", "")
sentences = sentences.replace("que", "")
sentences = sentences.replace("melankos", "")
sentences = sentences.replace("jogmfs", "")
sentences = sentences.replace("undipmenfess", "")
sentences = sentences.replace("ngalamfess", "")
sentences = sentences.replace("ubsansfess", "")
sentences = sentences.replace("karawangfess", "")
sentences = sentences.replace("stanmenfess", "")
sentences = sentences.replace("schfess", "")
sentences = sentences.replace("utbkfess", "")
```

```python
sentences
```

```python
wc = WordCloud(
    width=1200,
    height=1200,
    background_color="white",
    stopwords=StopWordRemoverFactory().get_stop_words(),
    min_font_size=10
).generate(sentences)

plt.figure(figsize=(10, 10), constrained_layout=True, facecolor=None)
plt.imshow(wc)
plt.axis("off")
plt.tight_layout(pad=0)

plt.show()
```
